```python
#if std is VERY SMALL compared to mean
    #→ Data is tightly packed
    #→ Low variation
    #→ More consistent
#If std is MODERATE compared to mean
    #→ Data has noticeable spread
    #→ Some variation
#If std is CLOSE TO or GREATER than mean
    #→ Data is highly spread
    #→ Very high variation
    #→ Possible extreme values
#From standard deviation you can conclude:
    #How consistent the data is
    #Whether values are clustered or scattered
    #Whether extreme values may exist
    #Whether market/demand is stable or unpredictable
#Std ÷ Mean × 100 = Percentage variation
#Interpretation Rule:
    #0–10% → Very low variation (very stable)
    #10–30% → Low to moderate variation
    #30–50% → Moderate variation
    #50%+ → High variation
    #100%+ → Very high variation
```

```python
# how many users converted to investors:
query = """
select count(distinct u.email) as total_converted
from users u
inner join investors i
on u.email = i.email
where u.is_verified == '1';

"""
```

```python
pd.read_sql_query(query,conn)
```

```python
# do verified users have higher probaility of becoming investors compared to unverified
query =""" 
select u.is_verified, 
count(distinct i.email)*1.0/ count(distinct u.email) as investors_probality
from users u
left join investors i
on u.email = i.email
group by u.is_verified;
"""
```

```python
pd.read_sql_query(query,conn)
```

```python
# what is the yearly growth rate of users vs investors is investors growth keeping up with user growth
# Growth rate = (new value - old value)/ old value
query = """  
select 
    year,
    total_count,
    lag(total_count) over (order by year) as previous_year,
    total_count - lag(total_count) over (order by year) as difference,
    (total_count - lag(total_count) over (order by year)) * 1.0 /
    lag(total_count) over (order by year) as growth_rate
from
(
    select 
        strftime('%Y', created_at) as year,
        count(*) as total_count
    from users
    group by strftime('%Y', created_at)
) t;

"""
```

```python
pd.read_sql_query(query,conn)
```

```python
# Percentage of active users out of total users
query = """ 
select total_active_users*100.0/total_users as active_users_percentage
from
(select count(*) as total_users,
(select count(*) from users where is_active = 1) as total_active_users
from users);
"""
```

```python
pd.read_sql_query(query,conn)
```

```python
# Is there a drop-off between users and investors registartion  and where is the largest funnel gap 
query = """ 
select total_users, total_verified,total_users - total_verified as drop_off_1, total_investors,total_investors - total_verified as drop_off_2
from
(
    select 
        (select count(*) from users) as total_users,
        (select count(*) from users where is_verified = 1) as total_verified,
        (select count(*) from investors) as total_investors
) ;

"""
```

```python
pd.read_sql_query(query,conn)
```

```python
# what is the overall startup approval rate:
#idea: (status ='aproved')/count(startups) *100

query ="""
select total_startups,
total_approved, 
total_startups *100.0/total_approved as aprroval_rate
from (
        SELECT
        (select count(*)from startups)as total_startups,
        (select count(*) from startups where status = 'approved')as total_approved
);
"""
```

```python
pd.read_sql_query(query,conn)
```

```python
#What percentage of approved startups become fully funded
#idea: count(approved_amount_raise) >= count(funding goal)
query ="""
select 
fully_funded,
total_approved,
fully_funded * 100.0 / total_approved as percentage
from (
    select
    (select count(*) from startups where status='approved') as total_approved,
    (select count(*) 
     from startups 
     where status='approved' and amount_raised >= funding_goal) as fully_funded
);

"""
#none of the approved startups are fully funded
```

```python
pd.read_sql_query(query,conn)
```

```python
#what percentage of startups raise at least 50% of their funding goal
#idea 1: amount_raised = 0.5 *funding_goal
#idea 2: idea 1/ count(startups)

query = """
select idea_1, total_count_startups, idea_1*100.0/total_count_startups as percentage
from(
select
(select count(*) from startups where amount_raised >= 0.5 * funding_goal)as idea_1,
(select count(*) from startups)as total_count_startups
);
"""
```

```python
pd.read_sql_query(query,conn)
```

```python
#Do startups with lowerfunding goals have higher sucess rate:
#idea 1: Low funding goal startups (funding_goal < average funding_goal)
#idea 2: High funding goal startups (funding_goal ≥ average funding_goal)
#Success rate = successful startups ÷ total startups in group.
#Using status = 'approved' is not exactly the original “success”, but it allows you to calculate a meaningful rate with the data you have.

query = """  
select successful_lower_funding_startups,
       total_lower_funding_startups,
       successful_lower_funding_startups*100.0/total_lower_funding_startups as success_rate
from (
    select
    (select count(*)
     from startups
     where funding_goal < (select avg(funding_goal) from startups)
       and status = 'approved') as successful_lower_funding_startups,

    (select count(*)
     from startups
     where funding_goal < (select avg(funding_goal) from startups)
    ) as total_lower_funding_startups
) t;


"""
```

```python
pd.read_sql_query(query,conn)
```

```python
# what is the average funding sucess across all startups
#1. funding_sucess = sum(amount_raised) / sum(funding_goal )
#2.avg(funding_sucess) = it's misleading, because it treats tiny startups the same as huge ones.

query = """

select  
sum(amount_raised),
sum(funding_goal ),
sum(amount_raised)*100.0/ sum(funding_goal ) as funding_sucess
from startups

"""
```

```python
pd.read_sql_query(query,conn)
```

```python
# what is total platforms overall capital efficeny:

query = """

select  
sum(amount_raised),
sum(funding_goal ),
sum(amount_raised)*100.0/ sum(funding_goal ) as capital_efficency
from startups

"""
```

```python
pd.read_sql_query(query,conn)
```

```python
# is funding distribution concentarted among a smalll numbers
#1. if max is greater than min and avg gives ans?
#Add percentiles (10th, 90th) → see the spread
#Calculate top 10% share of total funding → if top 10% of startups hold a large fraction of total funding, then distribution is truly concentrated
query = """
select 
    min_funding,
    max_funding,
    avg_funding
from(
    select
         (select min(funding_goal) from startups) as min_funding,
         (select max(funding_goal) from startups) as max_funding,
         (select avg(funding_goal) from startups) as avg_funding
 
 );
"""
#The funding distribution is concentrated among a small number of startups. 
#The minimum funding is very small compared to the maximum, and the average is significantly lower than the maximum. 
#This shows that a few startups received disproportionately high funding compared to the majority.
```

```python
pd.read_sql_query(query,conn)
```

```python
#what proption of toatl capital raised comes from top 10% highest funded startups
#Count total startups → 4,000.

#Top 10% of startups → 10% of 4,000 = 400 startups.

#Sort by amount raised (descending) → ensures you pick the largest funded startups.

#Sum the amount raised of these top 400 startups → total funding contributed by top 10%.

#Divide by total amount raised by all startups → gives the proportion of total capital contributed by top 10%.

query = """
select 
    total_amount_raised,
    top_10_percent_capital,
    top_10_percent_capital * 100.0 / total_amount_raised as capital_share_top_10_percent
from (
    select
        (select sum(amount_raised)
         from startups) as total_amount_raised,

        (select sum(amount_raised)
         from (
                select amount_raised
                from startups
                order by amount_raised desc
                limit 400
              ) top_startups
        ) as top_10_percent_capital
) funding_analysis;

"""
#total capital raised is about 10.06 billion.
#top 10% of startups raised about 2.52 billion out of the total 10.06 billion.
#The top 10% highest funded startups raised about 25% of the total capital on the platform.
#indicating that funding is somewhat concentrated among a smaller group of startups.
#but the majority of funding is still distributed across the remaining startups.
```

```python
pd.read_sql_query(query,conn)
```

```python
# what is the biggest drop-off in the ecosystem funnel

query = """
with funnel as (

    select 'users' as stage, count(*) as users, 1 as stage_order
    from users

    union all

    select 'investors' as stage, count(*) as users, 2
    from investors

    union all

    select 'investors_with_investment' as stage, count(distinct investor_email), 3
    from investments

)

select
    stage,
    users,
    lag(users) over(order by stage_order) as previous_step,
    (lag(users) over(order by stage_order) - users) * 100.0 /
    lag(users) over(order by stage_order) as drop_off_percentage
from funnel
order by stage_order;

"""
```

```python
pd.read_sql_query(query,conn)
```

```python
# is the platform improving overtime in terms of funding sucess rate and startup approval rate
# startup_approval_rate = approved startups / total startups 
# funding_sucess_rate = sum(amount_raised) / sum(funding_goal)
query = """

select year,
approved_startups,
funding_sucess_rate
from(
    select

(select strftime('%Y', created_at) from startups
) as year,

(select count(*) from startups) as total_startups,

(select 
count(status) 
from startups
where status = 'approved'
group by strftime('%Y', created_at)) as approved_startups,

(select sum(amount_raised*100.0)/sum(funding_goal) from startups group by strftime('%Y', created_at)) as funding_sucess_rate
);
"""
#every value in the SELECT clause is coming from scalar subqueries, and SQLite collapses them into a single result.
```

```python
pd.read_sql_query(query,conn)
```

```python
query = """
select 
strftime('%Y', created_at) as year,
sum(case when status='approved' then 1 else 0 end)*1.0 / count(*) *100.0 as startup_approval_rate,
sum(amount_raised)*100.0 / sum(funding_goal) as funding_sucess_rate
from startups
group by strftime('%Y', created_at);
"""
```

```python
pd.read_sql_query(query,conn)
```

```python
query = """  

select 
strftime('%Y', created_at) as year,

(
select count(*)
from startups s2
where s2.status = 'approved'
and strftime('%Y', s2.created_at) = strftime('%Y', s1.created_at)
) * 100.0 /

(
select count(*)
from startups s3
where strftime('%Y', s3.created_at) = strftime('%Y', s1.created_at)
) as startup_approval_rate,

(
select sum(amount_raised)*100.0 / sum(funding_goal)
from startups s4
where strftime('%Y', s4.created_at) = strftime('%Y', s1.created_at)
) as funding_sucess_rate

from startups s1
group by strftime('%Y', created_at)
order by year;

"""
```
